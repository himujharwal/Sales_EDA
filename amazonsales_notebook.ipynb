```python
new_df['rating'] = new_df['rating'].astype(str)

# Identify invalid entries by checking for conversion to float
def is_valid_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False
```

```python
invalid_entries = df[~df['rating'].apply(is_valid_float)]
invalid_entries.head()
# now exclude invalid entry
valid_entries = df[df['rating'].apply(is_valid_float)]
valid_entries.head()
new_df = valid_entries
```

```python
# Columns to clean
currency_columns = ['discounted_price', 'actual_price']  # Verify column names here
percentage_columns = ['discount_percentage']

# Check if 'new_df' exists and contains the required columns
print(new_df.columns)  # Print columns to check if 'discounted_price' is present

# Remove '₹' sign from currency columns
for col in currency_columns:
    if col in new_df.columns:  # Check if column exists before processing
        new_df[col] = new_df[col].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
    else:
        print(f"Column '{col}' not found in DataFrame.")

# Remove '%' sign from percentage columns
for col in percentage_columns:
    if col in new_df.columns:  # Check if column exists before processing
        new_df[col] = new_df[col].astype(str).str.replace('%', '', regex=False)
    else:
        print(f"Column '{col}' not found in DataFrame.")

# ... rest of your code ...
# rating

# remove , from rating count
new_df['rating_count'] = new_df['rating_count'].astype(str).str.replace(',', '', regex=False)

# remove nan value from
new_df = new_df.dropna()
```

```python
# Define the dictionary for data type conversion
dt_dict = {
    'discounted_price': 'float64',
    'actual_price': 'float64',
    'discount_percentage': 'float64',
    'rating': 'float64',
    'rating_count': 'float64' # Change data type to float64 to handle potential 'nan' strings
}

# Convert 'nan' strings to actual NaN values
new_df.replace('nan', np.nan, inplace=True)

# Remove rows with NaN values
new_df = new_df.dropna()

# Convert the data types
new_df = new_df.astype(dt_dict)

# Now convert 'rating_count' to integers, after handling NaNs
new_df['rating_count'] = new_df['rating_count'].astype('int64')

# Check the result

print(new_df.dtypes)
```

### Q1: What is the average rating for each product category?

```python
# Answer 1
average_rating_categorywise = new_df.groupby('category')['rating'].mean().reset_index()
average_rating_categorywise.sort_values(by='rating', ascending=False, inplace=True)
print(f"overall average rating : {average_rating_categorywise['rating'].median()}")

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='category', y='rating', data=average_rating_categorywise.head(10), order = average_rating_categorywise['category'].head(10))
plt.title('Top 10 Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')

# adding text inside of bar
for bar,rating in zip(barplot.patches, average_rating_categorywise['rating'].head(10)):
  barplot.text(


               bar.get_x() + bar.get_width() / 2,
      bar.get_height() - bar.get_height() * 0.2,
      rating,
      ha='center',
      va='center',
      rotation=45,
      size=10,
      color='black'
  )


```

### most product categories have generally positive customer feedback, with average ratings above **4.14**. However, some categories  have lower ratings, suggesting potential areas for improvement

### Q3: What is the distribution of discounted prices vs. actual prices?

```python
# this gives us polynomial  with scattorplot

sns.regplot(x='discounted_price', y='actual_price', data=new_df, order=2, ci=None, line_kws={"color": "red", "linewidth": 2})
plt.title('Discounted Price vs. Actual Price (Order 2)')
plt.xlabel('Discounted Price')
plt.ylabel('Actual Price')
plt.show();

print(f"discount_price_median: {np.median(new_df['discounted_price'])}")
print(f"actual_price_median: {np.median(new_df['actual_price'])}")


# Creating a KDE plot
sns.jointplot(x='discounted_price', y='actual_price', data=new_df, kind='kde')

plt.suptitle('KDE Plot of Discounted Price vs. Actual Price')
plt.xlabel('Discounted Price')
plt.ylabel('Actual Price')
plt.show()


```

### Q4: How does the average discount percentage vary across categories?

```python
avg_discount_per_category = new_df.groupby('category')['discount_percentage'].mean().reset_index()
avg_discount_per_category.sort_values(by='discount_percentage', ascending=False, inplace=True)
print(avg_discount_per_category.head(10))

# here are top 25 category to get clear graph
sns.barplot(x='category', y='discount_percentage', data=avg_discount_per_category.head(25), order = avg_discount_per_category['category'].head(25))
plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.show()
```

### Q5: What are the most popular product name?

```python


# Group by 'product_name' and count occurrences
product_counts = new_df['product_name'].value_counts().reset_index()
product_counts.columns = ['product_name_encoded', 'count']

# Sort by count in descending order
product_counts = product_counts.sort_values('count', ascending=False)

# Get the top 10 most popular products (encoded)
top_10_encoded = product_counts.head(10)

# Map encoded product names back to original names using the 'mapping' dictionary
top_10_names = top_10_encoded['product_name_encoded'].map(mapping['product_name'])

# Add the original names to the DataFrame
top_10_encoded['product_name'] = top_10_names

# Print the result
print(top_10_encoded)

# Plot the top 10 most popular products
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='product_name', data=top_10_encoded)
plt.title('Top 10 Most Popular Products')
plt.xlabel('Count')
plt.ylabel('Product Name')
plt.show()
```

### Q8: What is the correlation between discounted_price and rating?

```python


# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
correlation_coefficient_matrix = new_df[["discounted_price", "rating"]].corr()

sns.heatmap(correlation_coefficient_matrix, annot=True, cmap='coolwarm')
plt.title('Discounted Price vs. Rating Correlation')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
```

### Discounted price and rating have a weak positive correlation. This means that products with higher discounted prices tend to have **slightly** higher ratings, but the relationship is **not very strong**.
